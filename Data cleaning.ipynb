```python
import pandas as pd
import datetime
```

```python
aapl = pd.read_excel("AAPL null remove.xlsx", engine="openpyxl")
```

```python
aapl
```

```python
weekly_data = aapl.resample('W-Mon', label='right', closed = 'right', on='Date').mean().reset_index().sort_values(by='Date')
```

```python
weekly_data
```

```python
writer = pd.ExcelWriter('weekly.xlsx', engine='xlsxwriter')
```

```python
weekly_data.to_excel(writer, sheet_name='Sheet1')
```

```python
writer.save()
```
